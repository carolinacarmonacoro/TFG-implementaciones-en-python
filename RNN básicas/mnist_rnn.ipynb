```python
#Entrenamiento de una RNN para clasificar imГЎgenes del MNIST
#Evidentemente serГ­a mejor una red convolucional pero al ser un ejemplo sencillo y que ya he trabajado, lo hago con RNN.
#Cada imagen se tratarГЎ como una secuencia de 28 filas de 28 pГ­xeles cada una.
#Se usarГЎn capas de 150 neuronas recurrentes, que procesarГЎn las 28 filas secuencialmente, manteniendo una memoria de las filas anteriores.
#AdemГЎs se usarГЎ una capa de salida con 10 neuronas (una por clase) que se conectarГЎ al estado final, seguida de una capa softmax.
import tensorflow as tf
from tensorflow.keras import layers, models
```

```python
#Definimos los valores que queremos para la red y cargamos los datos del MNIST.
n_steps = 28
n_inputs = 28
n_neurons = 150
n_outputs = 10
learning_rate = 0.001
n_epochs = 20
batch_size = 150
mnist = tf.keras.datasets.mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
# Normalizamos los datos (imГЎgenes)
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

#Construimos el modelo
model = models.Sequential([
    #Solo nos quedamos con el Гєltimo estado (return_sequences=False por defecto)
    layers.SimpleRNN(n_neurons, input_shape=(n_steps, n_inputs)),
    #Capa de salida, por defecto activation=None (porque vamos a aplicar softmax despuГ©s)
    layers.Dense(n_outputs)
])

#Ahora hay que definir la pГ©rdida, el optimizador y las mГ©tricas.
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),     #Le aplica la softmax
    metrics=['accuracy']
)

#Una vez construida la red, la entrenamos.
model.fit(
    X_train, 
    y_train, 
    epochs=n_epochs, 
    batch_size=batch_size,
    validation_split=0.2,
    verbose=1
)

#EvaluaciГіn final
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=1)
print(f"\nPrecisiГіn final en test: {test_acc:.4f}")
```
